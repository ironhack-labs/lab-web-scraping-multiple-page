```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Python'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
links = soup.find_all('a')
```

```python
soup.find_all('a')
```

```python
links
```

```python
urls = []

for link in soup.find_all('a'):
    urls.append(link.get('href'))
```

```python
urls
```

```python
# Cleaning the urls

urls_df = pd.DataFrame({'Links' : urls})
              
```

```python
urls_df = urls_df.dropna()
```

```python
urls_df.head(50)
```

```python
urls_df['Links'] = urls_df['Links'][~urls_df['Links'].str.contains('#')]
```

```python
urls_df = urls_df.dropna()
```

```python
urls_df.tail(50)
```

```python
# correcting the internal links

for a, link in urls_df['Links'].items():
    if 'http' not in str(link):
        urls_df.at[a, 'Links'] = 'https://en.wikipedia.org' + str(link)
```

```python
# Find the number of titles that have changed in the United States Code since its last release point
```

```python
url3 = 'http://uscode.house.gov/download/download.shtml'
response3 = requests.get(url3)
response3
```

```python
soup3 = BeautifulSoup(response3.content, 'html.parser')
```

```python
titles_chaged = soup3.find_all('div', class_ = 'usctitlechanged')
```

```python
titles_final = [title.get_text() for title in titles_chaged]
titles_final = [title.strip() for title in titles_final]
titles_final = [title.replace('٭', '') for title in titles_final]
```
