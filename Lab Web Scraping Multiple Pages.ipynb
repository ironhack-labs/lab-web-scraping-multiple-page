```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#List all language names and number of related articles in the order they appear in wikipedia.org: url = 'https://www.wikipedia.org/'
```

```python
url = 'https://www.wikipedia.org/'
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
langs = soup.select('a.link-box > strong')

languages = []

for lan in langs:
    languages.append(lan.getText().strip())
languages
```

```python
nums = soup.select('a.link-box > small')

numbers = []

for num in nums:
    numbers.append(num.getText().strip())
numbers
```

```python
import re

numbers = [
    '6,847,000+ articles',
    '1,421,000+ 記事',
    '2.924.000+ Artikel',
    '1\xa0987\xa0000+ статей',
    '1.965.000+ artículos',
    '2\u202f621\u202f000+ articles',
    '1,429,000+ 条目 / 條目',
    '1.871.000+ voci',
    '۱٬۰۰۶٬۰۰۰+ مقاله',
    '1.128.000+ artigos'
]

def extract_number(string):

    number = re.sub(r'\D', '', string)
 
    return int(number)

numbers = [extract_number(item) for item in numbers]

print(numbers)
```

```python
languages = pd.DataFrame({
    'Languages': languages,
    'Numbers': numbers
})
languages
```
