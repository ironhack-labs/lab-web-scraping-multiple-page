```python
url_3 = 'https://www.fbi.gov/wanted/topten'
```

```python
response_3 = requests.get(url_3)
response_3
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
```

```python
response_3_1 = requests.get(url_3, headers=headers)
response_3_1
```

```python
soup_3 = BeautifulSoup(response_3_1.content, 'html.parser')
soup_3
```

```python
names = soup_3.find_all('h3', class_='title')
names
```

```python
wanted_names = [name.text.strip() for name in names]
wanted_names
```

```python
print("Top Ten FBI's Most Wanted Names:")
for index, name in enumerate(wanted_names, start=1):
    print(f"{index}. {name}")
```
