```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from random import randint
from time import sleep
```

```python
url = "https://www.joe.ie/music/the-50-most-iconic-songs-of-all-time-have-been-revealed-513205"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
print(soup)
```

```python
soup.select('body > div.main-wrapper.container.article-mobile-padding > main > section > div > div > article > div.article-content.p402_premium > ol > li:nth-child(1)')
```

```python
soup.select("div.article-content.p402_premium > ol > li:nth-child(1)")[0].text.strip()
```

```python
top_musics=[]
for i in range(1,50):
    url="div.article-content.p402_premium > ol > li:nth-child("+str(i)+")"
    top_musics.append(soup.select(url)[0].text.strip())
    
top_musics
```
