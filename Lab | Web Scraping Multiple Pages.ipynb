```python
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
url = "https://en.wikipedia.org/wiki/List_of_songs_recorded_by_Queen"
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content,"html.parser")
```

```python
soup
```

```python
soup.find_all('th', {'scope': 'row'})
```

```python
songs_queen = []
for row in soup.find_all('tr'):
    th_tag = row.find('th', {'scope': 'row'})
    if th_tag:
        song_name = th_tag.text.strip()
        songs_queen.append(song_name)
songs_queen
```

```python
url = "https://acharts.co/us_singles_top_100"
response = requests.get(url)
response.status_code
soup_us = BeautifulSoup(response.content,"html.parser")
```

```python
soup_us
```

```python
soup_us.find_all('span')
```

```python
song_US_top_100 = []
for row in soup_us.find_all('h3'):
    th_tag = row.find('span', {'itemprop': 'name'})
    if th_tag:
        song_name = th_tag.text.strip()
        song_US_top_100.append(song_name)
song_US_top_100
```

```python
url = "https://www.billboard.com/charts/hot-100/"
response = requests.get(url)
response.status_code
soup_us = BeautifulSoup(response.content,"html.parser")
```

```python
songs_billboard = []
for row in soup_us.find_all('li'):
    th_tag = row.find('h3', {'id': 'title-of-a-story'})
    if th_tag:
        song_name = th_tag.text.strip()
        songs_billboard.append(song_name)
songs_billboard
```

```python
hot_songs = song_US_top_100 + songs_queen + songs_billboard
```

```python
def check_song(song, hot_songs):
    if song in hot_songs:
        print("The song is present in the TOP.")
    else:
        print("The song is not present in the TOP.")
```

```python
song = input("Enter the song name: ")
check_song(song, hot_songs)
```
