```python
# Libaries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

# Exercise 1

```python
url = 'https://data.gov.uk/'
response = requests.get(url)
response
```

```python
'''
Exercise 1:
A list with the different kind of datasets available
'''
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
titles =soup.find_all('h3', class_ = 'govuk-heading-s dgu-topics__heading')
titles
```

```python
topics = [i.text.strip() for i in titles]
topics
```

```python
topics = []

# 2. Populate lists

for i in soup.find_all('h3', class_ = 'govuk-heading-s dgu-topics__heading'):
    topics.append(i.text.strip())
print(topics)

# 3. Create dataframe
df = pd.DataFrame({
    'Topic': topics
})
df
```

# Exercise 2

```python
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
response = requests.get(url)
response
```

```python
'''
Exercise 2:
Display the top 10 languages by number of native speakers stored in a pandas dataframe
'''
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
lang =soup.find_all('a', class_ = 'mw-redirect')
lang
```

```python
language = []

# 2. Populate lists

for i in soup.find_all('a', class_ = 'mw-redirect'):
    language.append(i.text.strip())
print(language)

# 3. Create dataframe
df = pd.DataFrame({
    'Language': language
})
df
```

```python
df1 = df.iloc[3:13].reset_index(drop=True)
df1
```

```python
num =soup.find_all('td')
num
```

```python
nums = [i.text.strip() for i in num]
nums
```

```python
def is_number(s):
    try:
        float(s)  # Try converting to float
        return True
    except ValueError:
        return False

# Filter the list to keep only numeric values
numeric_values = [s for s in nums if is_number(s)]
numeric_values
```

```python
df2 = pd.DataFrame({
    'Amount': numeric_values})
df2
```

```python
df3 = df2.iloc[:10]
df3
```

```python
df_final = pd.concat([df1, df3], axis = 1).reset_index(drop = True)
df_final
```
