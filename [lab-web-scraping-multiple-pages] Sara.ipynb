##### Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page

https://en.wikipedia.org/wiki/Python

https://en.wikipedia.org/wiki/Monty_Python

```python
# Creating the varible URL
url = 'https://en.wikipedia.org/wiki/Monty_Python'

# Geting the response
response = requests.get(url)
response
```

```python
# Getting the soup

from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify)
```

```python
links = soup.find_all('a')
links
```

```python
link_list = []
for link in links:
    href = link.get('href')
    if href and href.startswith('/wiki/'):
        link_list.append(href)

for link in link_list:
    print(f"https://en.wikipedia.org{link}")

else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

###### List all language names and number of related articles in the order they appear in wikipedia.org

https://www.wikipedia.org/

```python
# Creating the varible URL
url = 'https://www.wikipedia.org/'

# Geting the response
response = requests.get(url)
response
```

```python
# Getting the soup

from bs4 import BeautifulSoup

soupwiki = BeautifulSoup(response.content, 'html.parser')
print(soupwiki.prettify)
```

```python
language = soupwiki.find_all('div', class_ = 'central-featured-lang')
language
```

```python
language_list = []

language_list = [language.text.strip() for language in language]
```

```python
language_list
```

```python
language_names = []

for lang in language:
    lang_name = lang.find('strong').get_text(strip=True)
    language_names.append(lang_name)

print(language_names)
```

```python
article_counts = []

for lang in language:
    # Find the element that contains the article count
    article_count_element = lang.find('small')
    
    if article_count_element:
        # Extract the text within the element
        article_count = article_count_element.get_text(strip=True)
        article_counts.append(article_count)

# Print the list of article counts
print(article_counts)
```

```python
import re

# Sampling list of languages and article counts
languages_and_counts = ['English\n6,777,000+ articles',
 'Русский\n1\xa0961\xa0000+ статей',
 '日本語\n1,401,000+ 記事',
 'Español\n1.927.000+ artículos',
 'Deutsch\n2,877,000+ Artikel',
 'Français\n2\u202f587\u202f000+ articles',
 'Italiano\n1,845,000+ voci',
 '中文\n1,401,000+ 条目 / 條目',
 'العربية\n١٬٢٢٦٬٠٠٠+ مقالة',
 'فارسی\n۹۹۱٬۰۰۰+ مقاله']

# Extracting numbers using regular expressions and remove commas and other non-numeric characters
numbers = []
for item in languages_and_counts:
    match = re.search(r'(\d[\d,]*)', item)
    if match:
        number = re.sub(r'[^\d]', '', match.group(1))
        numbers.append(number)
    else:
        numbers.append(None)  # None for cases where no number is found

# Extracting language names (assuming language names are lines before numbers)
language_names = [line.split('\n')[0] for line in languages_and_counts]
```

```python
# Creating a DataFrame
df = pd.DataFrame({'Language': language_names, 'Article Numbers': numbers})

# Print the DataFrame
print(df)
```
