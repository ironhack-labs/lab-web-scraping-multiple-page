```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_songs_recorded_by_Miley_Cyrus')
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup.find_all('th')
```

```python
import pandas as pd

songs_miley = []  # Initialize an empty list

i = 1
r = 8
while i <= 100:
    song_name = soup.select('th')[r].text.strip()
    songs_miley.append(song_name)
    print(i, song_name)
    i += 1
    r += 1

Songs_miley = pd.DataFrame(songs_miley, columns=['Song Name'])
print(Songs_miley)
```

```python
#Find the number of titles that have changed in the United States Code since its last release point: url = 
# 'http://uscode.house.gov/download/download.shtml'
```

```python
changed_title_divs = soup5.find_all('div', class_='usctitlechanged')

changed_titles = []
for div in changed_title_divs:
    title_text = div.get_text().strip()
    changed_titles.append(title_text)

changed_titles = pd.DataFrame(changed_titles)
print(changed_titles)
```

```python
#Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe:
# url = 'https://www.emsc-csem.org/Earthquake/'
```

```python
response7 = requests.get('https://www.emsc-csem.org/Earthquake')
soup7 = BeautifulSoup(response7.text, "html.parser")

rows = soup7.find_all('tr')

for row in rows:
    print(row.get_text())
```

```python
earthquakes = []

rows = soup7.find_all('tr')[1:]  # Exclude the first row as it usually contains table headers

for row in rows:
    earthquake_info = [cell.get_text() for cell in row.find_all('td')]
    if len(earthquake_info) == 13:  # Adjust the number of columns if necessary
        earthquakes.append(earthquake_info)

last_20_earthquakes = earthquakes[:20]  # Select the last 20 earthquakes from the list

df = pd.DataFrame(last_20_earthquakes, columns=['Date', 'Time', 'Latitude', 'Longitude', 'Depth', 'Magnitude', 'Location',
                                                 'Region', 'Status', 'Source', 'Details', 'Map', ''])
df = df.drop(columns=[''])  # Remove the empty column

df.head(20)
```

```python
#List all language names and number of related articles in the order they appear in wikipedia.org: url = 'https://www.wikipedia.org/'
```

```python
response8 = requests.get('https://www.wikipedia.org/')
soup8 = BeautifulSoup(response8.text, "html.parser")

languages = []

strong_tags = soup8.find_all('strong')

for tag in strong_tags:
    language_name = tag.get_text()
    articles_tag = tag.find_next_sibling('small')
    if articles_tag:
        articles_count = articles_tag.get_text()
    else:
        articles_count = 'N/A'
    languages.append((language_name, articles_count))

for language in languages:
    print(language[0], '-', language[1])
```

```python
#A list with the different kind of datasets available in data.gov.uk: url = 'https://data.gov.uk/'
response9 = requests.get('https://data.gov.uk/')
soup9 = BeautifulSoup(response9.text, 'html.parser')

datasets = []

dataset_tags = soup9.find_all('h3', class_='govuk-heading-s')

for tag in dataset_tags:
    dataset_name = tag.get_text(strip=True)
    datasets.append(dataset_name)

for dataset in datasets:
    print(dataset)
```

```python
#Display the top 10 languages by number of native speakers stored in a pandas dataframe: url = 
# 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'

response10 = requests.get('https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers')
soup10 = BeautifulSoup(response10.text, 'html.parser')

table = soup10.find('table', {'class': 'wikitable sortable'})

rows = table.find_all('tr')[1:11]  # Exclude header row and select the top 10 rows

data = []
for row in rows:
    columns = row.find_all('td')
    language = columns[1].text.strip()
    speakers = columns[2].text.strip()
    data.append((language, speakers))

df_languages = pd.DataFrame(data, columns=['Language', 'Speakers'])

print(df_languages)
```
