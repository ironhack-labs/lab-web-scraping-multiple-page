```python
# Create a Python list with the top ten FBI's Most Wanted names: url = 'https://www.fbi.gov/wanted/topten'
from time import sleep
from random import randint
```

```python
url="https://www.fbi.gov/wanted/topten"
response=requests.get(url)

soup=BeautifulSoup(response.text,"html.parser")
print(soup)
```

```python
soup.select("#query-results-0f737222c5054a81a120bce207b0446a > ul > li:nth-child(1) > h3 > a")
```

```python
soup.select(" li:nth-child(1) > h3")[0].text.strip()
```

```python
most_wanted=[]
for i in range(1,11):
    url=" li:nth-child("+str(i)+") > h3"
    most_wanted.append(soup.select(url)[0].text.strip())
    wait_time=randint(1,4)
    sleep(wait_time)
    print("slept",wait_time,"seconds")
    
most_wanted
```
