```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://rfm.sapo.pt/top25rfm"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

chart_items = soup.find_all("div", class_="top25ItemData")

songs = []

for item in chart_items:
    title = item.select_one("div.tAuthor")
    if title:
        song_title = title.get_text().strip()
        artist = item.select_one("div.tMusic").get_text().strip()
        songs.append({"title": song_title, "artist": artist})
        
for song in songs:
    print(song["title"], "by", song["artist"])
```

```python
df1 = pd.DataFrame(songs)
df1
```

```python
url = "https://megahits.sapo.pt/top-10-as-10"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

chart_items = soup.find_all("td", class_="pg-gr-li-txtbox2")

songs = []

for item in chart_items:
    title = item.select_one("td.pg-gr-li-np1")
    if title:
        song_title = title.get_text().strip()
        artist = item.select_one("td.pg-gr-li-tx2").get_text().strip()
        songs.append({"title": song_title, "artist": artist})
        
for song in songs:
    print(song["title"], "by", song["artist"])
```

```python
df2 = pd.DataFrame(songs)
df2
```

```python
url = "https://megahits.sapo.pt/top-10-as-10"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

chart_items = soup.find_all("div", class_="NZAU7CsuZsMeMQ8zYUu")

songs = []

for item in chart_items:
    title = item.select_one("div.Type__TypeElement-sc-goli3j-0 cvuJgi t_yrXoUO3qGsJS4Y6iXX standalone-ellipsis-one-line")
    if title:
        song_title = title.get_text().strip()
        artist = item.select_one("td.pg-gr-li-tx2").get_text().strip()
        songs.append({"title": song_title, "artist": artist})
        
for song in songs:
    print(song["title"], "by", song["artist"])
```

```python
df_top = pd.concat([df1, df2])
```
